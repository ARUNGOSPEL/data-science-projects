```python
df = pd.get_dummies(df,columns =['country','gender']).astype('float')
df.head(5)
```

```python
df.info()
```

```python
X= df.iloc[:, 0:10]
X.head(2)
```

```python
y=df.iloc[:,8]
y.head()
```

```python
 ## MODEL BUILDING
## train test split
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.30,random_state=0)
```

```python
X_train
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test  =scaler.transform(X_test)
```

```python
from sklearn.model_selection import KFold,cross_val_score,StratifiedKFold
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression

from sklearn.metrics import accuracy_score
```

```python
knn_model = KNeighborsClassifier()
knn_model.fit(X_train,y_train)
```

```python
y_pred = knn_model.predict(X_test)
score  = accuracy_score(y_test,y_pred)*100
score
# print(np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape(len(y_test),1)),1))
```

```python
dtc_model = DecisionTreeClassifier()
dtc_model.fit(X_train,y_train)
```

```python
y_pred = dtc_model.predict(X_test)
score  = accuracy_score(y_test,y_pred)*100
score
```

```python
rfc_model = RandomForestClassifier()
rfc_model.fit(X_train,y_train)
```

```python
y_pred = rfc_model.predict(X_test)
score  = accuracy_score(y_test,y_pred)*100
score
```

```python
svc_model = SVC()
svc_model.fit(X_train,y_train)
```

```python
y_pred = svc_model.predict(X_test)
score  = accuracy_score(y_test,y_pred)*100
score
```

```python
gnb_model = GaussianNB()
gnb_model.fit(X_train,y_train)
```

```python
y_pred = gnb_model.predict(X_test)
score  = accuracy_score(y_test,y_pred)*100
score
```

```python
lr_model = LogisticRegression()
lr_model.fit(X_train,y_train)
```

```python
y_pred = lr_model.predict(X_test)
score  = accuracy_score(y_test,y_pred)*100
score
```

```python
models= []

models.append(('RFC',RandomForestClassifier()))
models.append(('LR',LogisticRegression(solver='liblinear',multi_class='ovr')))
models.append(('KNN',KNeighborsClassifier()))
models.append(('GNB',GaussianNB()))
models.append(('SVM',SVC()))
models.append(('DTC',DecisionTreeClassifier()))

print(models)
```

```python
results= []
names  = []
res    = []

for name,model in models:
    kfold_result = StratifiedKFold(n_splits=10,random_state=None)
    cv_results   = cross_val_score(model,X_train,y_train,cv=kfold_result,
                                   scoring='accuracy',error_score='raise')
    
    results.append(cv_results)
    names.append(name)
    res.append(cv_results.mean())
    
    print('%s: %f' %(name,cv_results.mean()))

plt.ylim(.90,.99)
plt.bar(names,res,color='red',width=0.6)
plt.title('ALGORITHM COMPARISON')
plt.show()
```
