### Let us analyse the features individually 

#### For the category type feature we shall use the BAR PLOT & PIE CHART
#### For the numerical type feature we shall use the BOX PLOT & HISTOGRAM

```python
## use a function to stop repetitions

def categorical_multi_plot (dataset= df, x =None, xlabel=None , explode =None):

    fig, ax = plt.subplots(1,2 , figsize =(10,5),dpi=200)

    sns.countplot(data=dataset, x= x, ax =ax[0], order =df[x].value_counts(ascending = True).index) 

    df[x].value_counts(ascending = True).plot(kind = 'pie', autopct ='%1.1f%%' , ax=ax[1], explode = explode)

    ax[0].set_xlabel(xlabel = xlabel, size=10, font= 'DejaVu Sans')

    ax[0].bar_label(ax[0].containers[0], label_type = 'edge', size =10, fontname ='Ariel', color = "#7f7f7f")
    
    fig.text(0.5,1, f'{xlabel} distribution', size =16, font = 'DejaVu Sans', ha='center', va='center')
    
    plt.show()
```

# 4.2) ------MULTIVARIATE ANALYSIS-----

```python
def multivariate_plot(dataset= df, x=None, xlabel=None, hue=None, rotation =None, bar_label= True, legend=False):

    dataset =dataset.groupby(x).mean()['salary_in_usd'].sort_values().reset_index()

    fig , ax = plt.subplots(figsize=(15,5),dpi=200)
    sns.barplot(data=dataset, x=x , y='salary_in_usd', hue= hue, ci=None)
    
    plt.xlabel(xlabel=xlabel, size=15, fontname ='DejaVu Sans')
    plt.ylabel(ylabel='SALARY IN USD', size=15, fontname ='DejaVu Sans')
    plt.xticks(rotation = rotation, size=12)
    plt.yticks(size=10)
    
    plt.title (label= f'SALARIES BY {xlabel}' ,size=20)
    
    if legend== True:
        plt.legend()
    
    if bar_label== True:
        ax.bar_label(ax.containers[0], label_type = 'edge', size =10, fontname ='Ariel',
                                                 color = "#7f7f7f", rotation= 'horizontal')
```

## 4.2.1) EXPERIENCE LEVEL  vs  SALARY 

```python
multivariate_plot(x='experience_level', xlabel= 'EXPERIENCE LEVEL')
```

### Lets have a look at the difference between the mean salaries of the different levels

```python
salary_by_experience = df.groupby('experience_level').mean()['salary_in_usd'].sort_values()
salary_by_experience
```

```python

ent_mid_diff = round((salary_by_experience.loc['midlevel'] - salary_by_experience.loc['entrylevel']),2)

percent_diff_1 = round(ent_mid_diff / salary_by_experience.loc['entrylevel'] *100,2)

print(f'DIFFERENCE BETWEEN ENTRY AND MID LEVEL           : {ent_mid_diff} ({percent_diff_1} %)')


mid_sen_diff =  round((salary_by_experience.loc['seniorlevel'] - salary_by_experience.loc['midlevel']),2)

percent_diff_2 = round(mid_sen_diff / salary_by_experience.loc['midlevel'] *100,2)

print(f'DIFFERENCE BETWEEN MID AND SENIOR LEVEL          : {mid_sen_diff} ({percent_diff_2} %)')


sen_exp_diff =  round((salary_by_experience.loc['expertlevel'] - salary_by_experience.loc['seniorlevel']),2)

percent_diff_3 = round(sen_exp_diff / salary_by_experience.loc['seniorlevel'] *100,2)

print(f'DIFFERENCE BETWEEN SENIOR AND EXPERT LEVEL       : {sen_exp_diff} ({percent_diff_3} %)')
```

## Observations

**salary is positively correlated with the experience level**

**Employees with higher experience get higher salary**

**Possible Conclusion :** An employer gets higher salary through his/her experience.
salary hike from a entry level to mid level is around +42%.If he keeps working and gain experience 
his salary will increase drastically.
