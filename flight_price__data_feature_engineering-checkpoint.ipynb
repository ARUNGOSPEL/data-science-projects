## 3.) FEATURE ENGINEERING

```python
# we use str index [0]  [1]  [2] in order to access the date,month,year after splitting


df['date']  =df['Date_of_Journey'].str.split('/').str[0]
df['month'] =df['Date_of_Journey'].str.split('/').str[1]
df['year']  =df['Date_of_Journey'].str.split('/').str[2]
```

```python
df.head(2)
```

```python
# change dtype as int for date,month,year

df['date']  =df['date'].astype(int)
df['month'] =df['month'].astype(int)
df['year']  =df['year'].astype(int)
```

```python
# Let us drop the date of journey column

df.drop('Date_of_Journey',axis=1,inplace=True)
```

**Let extract  the features available in ARRIVAL TIME COLUMN**

```python
df['Arrival_Time']

# The time is at the 0the index position, so using str.split we can extract the time data
```

```python
# b= df['Arrival_Time'].str.split(' ').str[0]
# b
```

```python
df['Arrival_Time'] = df['Arrival_Time'].apply(lambda x: x.split(' ')[0])

df['Arrival_Time']
```

**Extract the hours & minutes from arrival time column**

```python
#df['arrival_hour'].apply(lambda x: x.split(':')[0])
#df['arrival_min'].apply(lambda x: x.split(':')[1])

df['arrival_hour'] = df['Arrival_Time'].str.split(':').str[0]
df['arrival_min'] = df['Arrival_Time'].str.split(':').str[1]

df.head(2)
```

```python
# change the type as integer

df['arrival_hour'] = df['arrival_hour'].astype(int)

df['arrival_min']  = df['arrival_min'].astype(int)
```

```python
df.drop('Arrival_Time',axis=1,inplace=True)
```

```python
df.info()
```

**Lets extract available features in the departure time column**

```python
df['dep_hour'] = df['Dep_Time'].str.split(':').str[0]

df['dep_min'] = df['Dep_Time'].str.split(':').str[1]
```

```python
df['dep_hour'] = df['dep_hour'].astype(int)
df['dep_min']  = df['dep_min'].astype(int)
```

```python
df.drop('Dep_Time',axis=1,inplace=True)
```

```python
df.info()
```

```python
df['Total_Stops'].unique()
```

```python
# a = df['Total_Stops'].str.split(' ').str[0]

# we have non_stop in the 0th index position,so it is better to map those values with integers
```

```python
# map total stops with numbers

df['Total_Stops'] = df['Total_Stops'].map({'non-stop': 0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4,'nan':1})
```

```python
# deal with the route feature ------> better ignore this feature

df.drop('Route',axis=1,inplace=True)
df.head(1)
```

### Lets extract the features in duration hour column

```python
df['Duration']
```

```python
# convert duration into minutes

df['duration_hour'] = df['Duration'].str.split(' ').str[0].str.split('h').str[0]

df['duration_min']  = df['Duration'].str.split(' ').str[1]
```

```python
df.info()
```

```python
# at info we found the duration hour & duration min is in object dtype
# convert the dtype of both

# df['duration_hour'] = df['duration_hour'].astype(int)

# we found incorrect data at duration hour, because it has '5m' also a flight cannot travel anywhere for 5 min,
# which is invalid so we candrop that feature wher we have duration hour is 5 minutes
```

```python
df[df['duration_hour']== '5m']
```

```python
df.drop(6474,axis=0,inplace=True)
df.drop(2660,axis=0,inplace=True)
```

```python
df['duration_hour'] = df['duration_hour'].astype(int)
```

```python
# convert hours into minutes so multiply by 60

df['duration_hrs_into_min'] = df['duration_hour']*60
df['duration_hrs_into_min']
```

```python
df['duration_min'] = df['Duration'].str.split(' ').str[1].str.split('m').str[0]

# we cant change the dtype into integer, bcos of float 'nan' in the column 
# so we hav to replace nan with 0 

# df['duration_min'] = df['duration_min'].replace(np.nan,0)

df['duration_min'] = df['duration_min'].fillna(0)
df['duration_min'].astype(int)
```

```python
df.head()
```

```python
df.drop('Duration',axis=1,inplace=True)
```
